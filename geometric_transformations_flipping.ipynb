```python
import cv2
from matplotlib import pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
image_path = '/content/drive/MyDrive/computervision/data/image.png'
image = cv2.imread(image_path)
plt.imshow(image)
```

```python
# vertically flip
flipped_image = cv2.flip(image, 0)
plt.imshow(flipped_image)
```

```python
#flip horizontally
flip_image= cv2.flip(image,1)
plt.imshow(flip_image)
```

```python
both_flip= cv2.flip(image,-1)
plt.imshow(both_flip)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

axs[0].imshow(image)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(flipped_image)
axs[1].set_title('Vertically Flipped Image')
axs[1].axis('off')

axs[2].imshow(flip_image)
axs[2].set_title('Horizontally Flipped Image')
axs[2].axis('off')

axs[3].imshow(both_flip)
axs[3].set_title('Both Flipped Image')
axs[3].axis('off')

plt.tight_layout()
plt.show()
```
